```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```


Splitting the data into training and test data

```python
# Splitting the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
# Checking the shapes of the datasets
print(X.shape, X_train.shape, X_test.shape)
```

Standardized the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()

X_train_std = scalar.fit_transform(X_train)
X_test_std = scalar.transform(X_test)
```

Building the neural networkdownload.png

```python
#importing Tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of Neural Network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
#compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#training the neural network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

Visualizing Accuracy and Loss

```python
# plotting the accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
# plotting the accuracy
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

Accuracy of the model on the test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

Building the predictive system

```python
input_data=(20.59,21.24,137.8,1320,0.1085,0.1644,0.2188,0.1121,0.1848,0.06222,0.5904,1.216,4.206,75.09,0.006666,0.02791,0.04062,0.01479,0.01117,0.003727,23.86,30.76,163.2,1760,0.1464,0.3597,0.5179,0.2113,0.248,0.08999)

# change the input_data to a numpy array
input_data_as_numpy_array=np.asarray(input_data)

#reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

#Standardizing the input data
input_std_data = scalar.transform(input_data_reshaped)

prediction = model.predict(input_std_data)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0]==0):
  print('The breast cancer is Malignant')
else:
  print('The breast cancer is Benign')
```
